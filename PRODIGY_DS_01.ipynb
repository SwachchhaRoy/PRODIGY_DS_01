## Population Dataset

```python
# importing all necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read and display data
data = pd.read_csv('E:\PRODIGY INFOTECH\Task1\API_SP.POP.TOTL_DS2_en_csv_v2_6508519.csv')
pd.DataFrame(data)
```

```python
# finding null values
data.isnull().sum()
```

```python
# since it is a population dataset, we can fill the null values with the previous non-null value along same column
data = data.fillna(method = 'ffill')
data.head()
```

```python
# checking whether any null values are there in any column 
data.isnull().sum()
```

```python
# checking uniqueness of the Country names
print(data['Country Name'].unique())
print('\n')
print('The total number of unique values in Country Name column is:',data['Country Name'].nunique())
```

```python
# checking uniqueness of the Country Codes
print(data['Country Code'].unique())
print('\n')
print('The number of unique Country code is:',data['Country Code'].nunique())
```

```python
data['Indicator Name'].nunique()
```

```python
data['Indicator Code'].nunique()
```

```python
# dropping unneccary columns Indicator Name and Indicator Code as it has same values and does'nt impact our dataset
data.drop(['Indicator Name','Indicator Code'], axis = 1,inplace = True)
data.head(5)
```

```python
data.columns
```

```python
# plotting the population of every year
columns = ['1960', '1961', '1962', '1963', '1964',
       '1965', '1966', '1967', '1968', '1969', '1970', '1971', '1972', '1973',
       '1974', '1975', '1976', '1977', '1978', '1979', '1980', '1981', '1982',
       '1983', '1984', '1985', '1986', '1987', '1988', '1989', '1990', '1991',
       '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000',
       '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009',
       '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018',
       '2019', '2020', '2021', '2022']

for i in columns:
    plt.hist(data[i], color = 'm', edgecolor = 'k')
    plt.xlabel(i)
    plt.show()
```

```python
# plotting horizontal bar population vs years
plt.figure(figsize = (20,20))
years = data.columns[2:]
values = data[years].sum()
plt.barh(years,values, color='blue',edgecolor = 'k')
plt.xlabel('Total Population',size = 15)
plt.ylabel('Years',size=15)
plt.title('Total Population per year',size = 20)
plt.show()
```

```python
# finding the maximum and minimum populations of each year
for i in data.columns[2:]:
    print(f'The maximum population of the year {i} is',data[i].max())
    print(f'The minimum population of the year {i} is',data[i].min())
    print('\n')
```
